Extract principal components

```python
def get_components(logmelspectrograms):
    # apply PCA to data
    pca = PCA(0.9)
    principalComponents = pca.fit_transform(logmelspectrograms)
    print(f"Number of extracted components {pca.n_components_}")
    return principalComponents
```

```python
pc = {}
for d in ds:
    print(f'For {d}')
    pc[d] = get_components(sgrams[d])
```

Compute per label matrix

```python
def compute_correlations(md, pc, t):
    corr = {}
    for d in ds:
        corr[d] = {}

        labels = md[d].label.unique()
        for l in labels:
            indeces_l = md[d][md[d]['label']==l].index.values.astype(int)
            data_l = pc[d][indeces_l]
            # get corr matrix
            m = pd.DataFrame(data_l).corr(method = t)
            # only upper triangular
            m = m.where(np.triu(np.ones(m.shape)).astype(np.bool))
            # remove elements in diagonal
            np.fill_diagonal(m.values, 'NaN')
            # stack array and get values
            m = m.stack().values
            corr[d][l] = m
    return corr
```

```python
corr_pearson = compute_correlations(md, pc, 'pearson')
corr_spearman = compute_correlations(md, pc, 'spearman')
corr_kendall = compute_correlations(md, pc, 'kendall')
```

Measure distance https://stackoverflow.com/questions/1401712/how-can-the-euclidean-distance-be-calculated-with-numpy

```python
def dist_calc(corr):
    dist = {}
    for d in corr:
        print(d)
        dist[d] = []
        for l1 in corr[d]:
            dist[d].append([np.linalg.norm(corr[d][l1]-corr[d][l2]) for l2 in corr[d]])
    return dist
```

```python
dist_pearson = dist_calc(corr_pearson)
dist_spearman = dist_calc(corr_spearman)
dist_kendall = dist_calc(corr_kendall)
```
